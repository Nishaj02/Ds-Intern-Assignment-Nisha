```python
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
# Preprocessing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("/content/Fraud DATAAAAAAAAAA.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.boxplot(x=df['amount'])
plt.show()
```

amount is highly right-skewed

Fraud transactions occur at high values

Outliers are important signals, so we do not remove them

```python
df['log_amount'] = np.log1p(df['amount'])
```

```python
#Multicolinearity handling
corr = df[['oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
#Feature Engineering to Reduce Multicollinearity
df['balanceDiffOrig'] = df['oldbalanceOrg'] - df['newbalanceOrig']
df['balanceDiffDest'] = df['newbalanceDest'] - df['oldbalanceDest']
```

Model Used: Random Forest Classifier
Because-
Handles non-linear relationships

Robust to outliers

Works well for imbalanced data

Provides feature importance

```python
le = LabelEncoder()
df['type_encoded'] = le.fit_transform(df['type'])
```

```python
features = [
    'type_encoded',
    'log_amount',
    'balanceDiffOrig',
    'balanceDiffDest',
    'oldbalanceOrg',
    'newbalanceDest'
]

X = df[features]
y = df['isFraud']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:,1]
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
roc_auc_score(y_test, y_prob)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr, label='ROC Curve')
plt.plot([0,1],[0,1],'--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

High ROC-AUC (~0.970.99)

Strong recall for fraud class

Suitable for real-world fraud detection

```python
importances = pd.Series(rf.feature_importances_, index=features)
importances.sort_values(ascending=False)
```
